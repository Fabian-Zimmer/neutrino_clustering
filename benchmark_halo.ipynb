```python
# To be read from sim box?
rho_0 = rho0_MW
r_s = Rs_MW
Rvir = Rvir_MW

# Total mass until r
def fNFW(arg):
    return np.log(1+arg) - (arg/(1+arg))

def M_in_r(r):
    return 4*Pi*rho_0*r_s**3*fNFW(r/r_s)

def T_in_r(r, Rvir):
    return M_in_r(r)/M_in_r(Rvir)


# Uniformly sample 235953 particles.
# (one of the MW like halos in the sim box has this many)

sample = np.random.uniform(size=100_000)

T_arr = T_in_r(sample, Rvir)

plt.plot(radii/kpc, T_arr)
plt.show()
```
